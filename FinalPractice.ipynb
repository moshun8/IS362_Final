```python
# Grab names and ID# of Bundesliga teams
bl_teams = '''SELECT t.team_long_name, t.team_api_id
            FROM Team as t
            JOIN Match as m
            ON t.team_api_id = m.away_team_api_id
            JOIN League as l
            ON m.league_id = l.id
            WHERE l.id = 7809 AND m.season = '2015/2016'
            ORDER BY t.team_long_name
            '''

bl_team_list = pd.read_sql_query(bl_teams, con=con).drop_duplicates()
bl_team_list.head(30)
```

```python
#EPL Team list
epl_teams = '''SELECT t.team_long_name, t.team_api_id
            FROM Team as t
            JOIN Match as m
            ON t.team_api_id = m.away_team_api_id
            JOIN League as l
            ON m.league_id = l.id
            WHERE l.id = 1729 AND m.season = '2015/2016'
            ORDER BY t.team_long_name
            '''

epl_team_list = pd.read_sql_query(epl_teams, con=con).drop_duplicates()
epl_team_list.head(30)
```
